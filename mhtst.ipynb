```python
import pandas as pd

df = pd.read_csv(r"C:\Users\NGUYEN PHUONG BICH\Downloads\heart_disease_uci.csv")
del df['id']
del df['dataset']
df
```

```python
df["sex"] = df["sex"].astype('string').apply(lambda x: 1 if x == "Male" else (0 if x == "Female" else -1))


df['cp'] = df['cp'].astype('string').apply(lambda x : 0 if  x == 'typical angina' else (1 if x == 'atypical angina' else (2 if x == 'non-anginal' else 3 if x == 'asymptomatic' else -1)))

df['fbs'] = df['fbs'].astype('string').apply(lambda x : 0 if x == 'FALSE' else (1 if x == 'TRUE' else -1))

df['restecg'] = df['restecg'].astype('string').apply(lambda x : 0 if x == 'normal' else (1 if x == 'st-t abnormality'else (2 if x == 'lv hypertrophy' else -1)))

df['exang'] = df['exang'].astype('string').apply(lambda x:0 if x == 'FALSE' else (1 if x == 'TRUE' else -1))

df['slope'] = df['slope'].astype('string').apply(lambda x : 0 if x == 'upsloping' else (1 if x == 'flat' else (2 if x == 'downsloping' else -1)))

df['thal'] = df['thal'].astype('string').apply(lambda x : 0 if x == 'normal' else (1 if x == 'fixed defect' else (2 if x == 'reversible' else (3 if x == 'not described' else -1))))
```
